```python
import glob
```

```python
def read_word_phn(phn_file):
    start = -1
    cur_end = 0
    size = 0
    for line in open(phn_file, encoding="iso-8859-15"):
        s,e,phone = line.strip().split(None,2)
        
        if len(phone) < 3:
            if size > 0:
                yield (cur_end-start, size)
                size = start = cur_end = 0
        else:
            if size == 0:
                start = int(s)
        size += 1
        cur_end = int(e)
```

```python
items = []
for phn in glob.glob("new_aligned/*.aligned"):
    for i, p in enumerate(read_word_phn(phn)):
        l=p[0]
        s=p[1]
        items.append((phn,i, l,s,l/s))
```
